```python
import pandas as pd
import datetime as datetime

#%% Definición del archivo a procesar
data = pd.read_csv('C:\\Users\\Adriana\\Documents\\logs_IFYP_Adriana.csv')
df2 = data[(data["Nombre completo del usuario"] == "GARCIA PADILLA ADRIANA DE GUADALUPE")]
df2

#%% Función data quality report
def dataqualityreport(data):
# Creamos un DataFrame con las columnas del archivo de alumnos
columns = pd.DataFrame(list(data.columns.values),
columns = ['Columns Names'],index=list(data.columns.values))
# lista de tipos de variables
data_types = pd.DataFrame(data.dtypes,
columns= ['Data Type'])
# Contando datos perdidos
missing_data_counts = pd.DataFrame(data.isnull().sum(),
columns = ['Missing Values'])
# Cuántos datos tenemos
present_data_counts = pd.DataFrame(data.count(),
columns=['Present Values'])
# Valores únicos
unique_value_counts = pd.DataFrame(columns=['Unique Values'])
for v in list(data.columns.values):
unique_value_counts.loc[v] = [data[v].nunique()]
# Valores mínimos por variable
minimum_values = pd.DataFrame(columns=['Minimum Values'])
for v in list(data.columns.values):
try:
minimum_values.loc[v]=[data[v].min()]
except:
pass
# Valores máximos por variable
maximum_values = pd.DataFrame(columns=['Maximum Values'])
for v in list(data.columns.values):
try:
maximum_values.loc[v]=[data[v].max()]
except:
pass
# Juntar la información que se logró recabar
data_quality_report = columns.join(data_types).join(missing_data_counts).join(present_data_counts).join(unique_value_counts).join(minimum_values).join(maximum_values)
return data_quality_report


#%% Data Quality Report
dqr = dataqualityreport(data) #ESTA ES LA LLAMADA A LA FUNCIÓN

#%% Cambiar la variable Hora a solamente la fecha para filtrar los datos de este semestre
# https://www.journaldev.com/23365/python-string-to-datetime-strptime

def change_fort_date(x):
return datetime.datetime.strptime(x,"%d/%m/%Y %H:%M")

new_date = data.Hora.apply(change_fort_date)

# Seleccionar los datos solamente de este semestre
indx = new_date>datetime.datetime.strptime('01/08/2018','%d/%m/%Y')

new_data = data.loc[indx]

# Excluyendo al profesor
idx2 = new_data['Nombre completo del usuario']!='GARCIA PADILLA ADRIANA DE GUADALUPE'
new_data = new_data.loc[idx2]

#%% 
dqr = dataqualityreport(new_data)
dqr

#%%
accesos_usuario = pd.value_counts(new_data['Nombre completo del usuario'])
accesos_usuario

#%%
eventos = pd.value_counts(new_data['Nombre evento'])
eventos

#%%
origen = pd.value_counts(new_data['Origen'])
def filtrar_origen(x):
x = str(x)
return x[0:23]

temp = new_data['Descripción'].apply(filtrar_origen)
origen = pd.value_counts(temp)
origen

#%%
# filtrar las IP por el dominio xxx.xxx
def filtrar_dominio(x):
x = str(x)
return x[0:7]

temp = new_data['Dirección IP'].apply(filtrar_dominio)
dir_Dom = pd.value_counts(temp)
dir_Dom

#%%
componente = pd.value_counts(new_data['Componente'])
componente

#%% filtrar horas
def filtrar_hora(x):
x = str(x)
temp = x.split()
return temp[1][0:2]

temp = new_data['Hora'].apply(filtrar_hora)
horas = pd.value_counts(temp)
horas
```
